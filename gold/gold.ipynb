```python
# import requests

# url = "https://rate.bot.com.tw/gold?Lang=zh-TW"

# html = requests.get(url)

# print(html.text)

# 使用urllib.request爬網
import urllib.request as request
from bs4 import BeautifulSoup as BS

url = "https://rate.bot.com.tw/gold?Lang=zh-TW"
with request.urlopen(url)as response:
    data = response.read().decode("utf-8") # 讀response並解碼成utf-8
    bs_data = BS(data,"html.parser")
    gold_in = (bs_data.find_all("td")[5].text.replace("回售","").strip())
    gold_out = (bs_data.find_all("td")[2].text.replace("買進","").strip())
    print("黃金回售:",gold_in)
    print("黃金買進:",gold_out)
    
```

```python
import requests
from bs4 import BeautifulSoup as BS

google_url = "https://www.google.com/search"

# 查詢參數
my_params = {"q":"奶子航線"}

html = requests.get(google_url,params=my_params)

# 確認是否請求成功
if html.status_code == 200:
    soup = BS(html.text,"html.parser")

    # 測試用
    # print(soup.prettify())
    # div_items = soup.find_all("span")
    # print(div_items)

    # items = soup.select('div.kCrYT > a[href^="/url"]')
    # 尋找搜尋貼文標題&url
    headline = soup.select("h3.zBAuLc")
    items_url = soup.select("span a")

    # 刪除最後一筆資料
    # 因為我覺得怪怪的
    del headline[-1]
    del items_url[-1]

    # Debug用
    # print(items_url)

    # 輸出
    for i,j in zip(headline,items_url):
        print("標題: " + i.text)
        print("url:"+j.get('href').split("=")[1].replace(r"%3F","?").replace(r"%3D",'=').replace(r"%26","&"))

        print()
```
