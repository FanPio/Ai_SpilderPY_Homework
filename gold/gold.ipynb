```python
# import requests

# url = "https://rate.bot.com.tw/gold?Lang=zh-TW"

# html = requests.get(url)

# print(html.text)

# 使用urllib.request爬網
import urllib.request as request
from bs4 import BeautifulSoup as BS

url = "https://rate.bot.com.tw/gold?Lang=zh-TW"
with request.urlopen(url)as response:
    data = response.read().decode("utf-8") # 讀response並解碼成utf-8
    bs_data = BS(data,"html.parser")
    gold_in = (bs_data.find_all("td")[5].text.replace("回售","").strip())
    gold_out = (bs_data.find_all("td")[2].text.replace("買進","").strip())
    print("黃金回售:",gold_in)
    print("黃金買進:",gold_out)
    
```
